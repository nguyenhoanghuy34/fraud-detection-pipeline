## Problem Statement

### Yêu cầu bài toán

Xây dựng một mô hình Machine Learning có khả năng **phát hiện các giao dịch gian lận (Fraud Detection)** dựa trên thông tin lịch sử giao dịch.

Mục tiêu của mô hình là phân tích các mẫu giao dịch và xác định những giao dịch có nguy cơ gian lận, từ đó hỗ trợ phát hiện sớm và giảm thiểu thiệt hại tài chính.

---

### 1. Phân loại giao dịch

Bài toán được xây dựng dưới dạng **Binary Classification**, trong đó mô hình cần phân loại giao dịch thành hai nhóm:

- **Fraud**: Giao dịch gian lận
- **Legitimate / Normal**: Giao dịch hợp lệ

---

### 2. Phát hiện giao dịch đáng ngờ

Mô hình cần:

- Phân tích đặc điểm của các giao dịch trong quá khứ.
- Tìm ra các mẫu hành vi bất thường.
- Xác định các giao dịch có khả năng gian lận.
- Hỗ trợ ngăn chặn các giao dịch gây tổn thất.

---

### 3. Xử lý dữ liệu mất cân bằng (Imbalanced Data)

Một trong những thách thức lớn của bài toán Fraud Detection là dữ liệu thường bị mất cân bằng.

Trong thực tế:

- Giao dịch bình thường chiếm phần lớn dữ liệu.
- Giao dịch gian lận chỉ chiếm một tỷ lệ rất nhỏ.

Ví dụ:

99.8% Normal Transactions
0.2% Fraud Transactions

Vì vậy, Accuracy không phải là metric đánh giá phù hợp duy nhất, do mô hình có thể đạt Accuracy cao nhưng vẫn bỏ sót nhiều giao dịch gian lận.

---

### 4. Tối ưu khả năng phát hiện gian lận

Mô hình cần cân bằng giữa khả năng phát hiện fraud và hạn chế cảnh báo sai.

Các mục tiêu chính:

- **High Recall**
  - Phát hiện được nhiều giao dịch gian lận thực sự.
  - Giảm số lượng Fraud bị bỏ sót.

- **Good Precision**
  - Giảm số lượng giao dịch hợp lệ bị đánh nhầm là gian lận.

---

### 5. Evaluation Metrics

Các metric được sử dụng để đánh giá mô hình:

- Precision
- Recall
- F1-score
- ROC-AUC
- PR-AUC
- Confusion Matrix

---

### 6. Business Objective

Mục tiêu cuối cùng của hệ thống:

- Giảm **False Negative (FN)**:
  - Giao dịch gian lận nhưng mô hình dự đoán là hợp lệ.

- Duy trì sự cân bằng giữa:
  - Phát hiện chính xác giao dịch gian lận.
  - Tránh chặn nhầm khách hàng hợp lệ.

Hệ thống Fraud Detection hiệu quả giúp tăng độ an toàn giao dịch và giảm rủi ro tài chính.
